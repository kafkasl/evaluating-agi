# Evaluating AGI: BenchPress Scaffold


A scaffold to test new AGI benchmarks using the `benchpress` matrix completion method. 

When proposing a new benchmark (especially in competitions), it's hard to know if it's actually measuring something **novel**, or just redundant with existing tests (like AIME or MMLU). This scaffold uses the [BenchPress](https://open.substack.com/pub/dimitrisp/p/you-dont-need-to-run-every-eval) methodology to predict how models *should* score based on the entire ecosystem of existing benchmarks. 

* If your new benchmark's scores can be perfectly predicted, it's highly correlated with existing tests (redundant). 
* If they can't be predicted, you've found a novel axis of evaluation (or it's just noise).

This toolkit provides a quick, mathematical way to check and defend the novelty of your benchmarks.

## Usage

The core of the library is the `check_novelty` function. It calculates a **Leave-One-Out (LOO) median error** to test if your benchmark is novel.

*   `~3 pts median error`: **Redundant** (Highly predictable from other benchmarks)
*   `~8 pts median error`: **Somewhat Novel**
*   `~12+ pts median error`: **Novel** (Measures something genuinely new)
*   `~18+ pts median error`: **Noise** (Random or broken benchmark)

````python
from evaluating_agi.benchpress import check_novelty

# 1. Provide the scores your benchmark assigned to various models.
#    (Keys must match the valid model IDs in the BenchPress matrix)
my_benchmark_scores = {
    'gpt-4.5': 85.0,
    'claude-3.7-sonnet': 82.5,
    'o3-mini-high': 88.0,
    'claude-opus-4': 78.0
}

# 2. Evaluate how novel these scores are against the existing AGI landscape
report = check_novelty(my_benchmark_scores, name='LogicPuzzle-100')
````
